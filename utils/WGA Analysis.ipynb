## Data Analysis

Using the [dataset](https://www.wga.hu/database/download/index.html) from the Web Gallery of Art, it was put in the dataframe below to show its contents. 

````python
table = pd.read_excel("catalog.xlsx")
table
````

````python
print(f"dataset length: {len(table)}")
````

**As you can see the dataset has 51459 rows and 11 columns representing the metadata of each art.**  

### Analysis by SCHOOL 

For the purpose of web scraping, the dataset appears to be very big therefore a subset will be used. This subset will be chosen based on the SCHOOL and FORM 

````python
listschool = table["SCHOOL"].unique()
dic = {}
for i in listschool:
    print (f"Number of {i} forms, {len(table[table.SCHOOL == i])}")
   
````

**As we can see the highest number of FORMS is from the Italian SCHOOL and the lowest number is from the Finnish SCHOOL.** 

````python
listschool = table["SCHOOL"].unique()  
for i in listschool:
    print (f"Number of {i} paintings, {len(table[(table.SCHOOL == i) & (table.FORM == 'painting')])}")
   
````

**Now it is filtered by the FORM "painting" and SCHOOL**

## French Paintings Dataset

Based on the analysis, I am creating my own dataset that I will be calling "French Paintings Dataset" the reason I chose the French **SCHOOL** over other schools, is because the amount of data related to the French SCHOOL and the **FORM** ```paintings``` is enough to be annotated and to be used as a machine learning model input.  

````python
french_paintings = table[(table.SCHOOL == "French") & (table.FORM == 'painting')].reset_index()
french_paintings
````

````python
french_paintings.to_csv("french_paintings_metadata.csv")
````
